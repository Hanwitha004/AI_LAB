```python
import heapq
import copy


goal_state = [[1, 2, 3],
               [8, 0, 4],
               [7, 6, 5]]


def find_zero(puzzle):
    for i in range(3):
        for j in range(3):
            if puzzle[i][j] == 0:
                return i, j


def calculate_manhattan_distance(puzzle, goal):
    distance = 0
    for i in range(3):
        for j in range(3):
            value = puzzle[i][j]
            if value != 0:

                for goal_i in range(3):
                    for goal_j in range(3):
                        if goal[goal_i][goal_j] == value:
                            distance += abs(i - goal_i) + abs(j - goal_j)
                            break
                    else:
                        continue
                    break
    return distance


def generate_neighbors(puzzle):
    neighbors = []
    x, y = find_zero(puzzle)


    moves = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    for move in moves:
        new_x, new_y = x + move[0], y + move[1]
        if 0 <= new_x < 3 and 0 <= new_y < 3:
            new_puzzle = copy.deepcopy(puzzle)
            new_puzzle[x][y], new_puzzle[new_x][new_y] = new_puzzle[new_x][new_y], new_puzzle[x][y]
            neighbors.append(new_puzzle)

    return neighbors


def a_star_8_puzzle_manhattan(start, goal):

    priority_queue = []
    heapq.heappush(priority_queue, (0, 0, start))

    visited = set()
    visited.add(tuple(tuple(row) for row in start))

    while priority_queue:

        f, g, current_puzzle = heapq.heappop(priority_queue)


        h = calculate_manhattan_distance(current_puzzle, goal)

        print(f"Current Puzzle:\n{current_puzzle}\n g: {g}, h: {h}, f: {f}")


        if current_puzzle == goal:
            return g


        for neighbor in generate_neighbors(current_puzzle):
            if tuple(tuple(row) for row in neighbor) not in visited:
                visited.add(tuple(tuple(row) for row in neighbor))
                new_g = g + 1
                new_h = calculate_manhattan_distance(neighbor, goal)
                new_f = new_g + new_h
                heapq.heappush(priority_queue, (new_f, new_g, neighbor))

    return -1


start_state = [[2, 8, 3],
               [1, 6, 4],
               [7, 0, 5]]


depth_of_solution = a_star_8_puzzle_manhattan(start_state, goal_state)
print(f"Depth of solution: {depth_of_solution}")
```
